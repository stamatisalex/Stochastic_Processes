Αρχικά ορίζουμε 2 συναρτήσεις που υλοποιούν έναν απλό συμμετρικό περίπτατο στον $Ζ$  και στον $Ζ^2$ . Ο κώδικας υλοποίησης αυτών είναι ο εξής:

```python
import numpy as np
import statistics as stat
from numpy.random import choice

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)  # default figure size
```

## Απλός Συμμετρικός περίπατος στον $\mathbb{Z}$

```python
np.random.seed(2021)  # same results are previous run

# random walk generator:
def rand_walk_Z(start = 0, probs = (0.5, 0.5)):
    x = start
    steps = [-1, 1]  # probs order should be (left, right)
    while True:  # continue for ever in needed
        x += choice(steps, p=probs)
        yield x  # it's a generator!
```

## Τυχαίος περίπατος στον $\mathbb{Z}^2$

```python
# random walk in Z^2 generator:
def rand_walk_Z2(start=0, probs=[1/3,2/3]):
    x = start
    while True:  
        if (x == start):
            x += np.random.choice([1, 0], p=probs)
        else:
            x += np.random.choice([1,-1], p=probs)
        yield x

```

## Άσκηση 1

Για τον απλό τυχαίο περίπατο, μέχρι αυτός να βγει από το $[-70, 100]$ προσομοιώστε:

* 500 διαδρομές με τάση προς τα αριστερά $p(k, k+1) = \frac{1}{5}, p(k,k-1) = \frac{4}{5}$
* 500 διαδρομές με τάση προς τα δεξιά $p(k, k+1) = \frac{4}{5}, p(k,k-1) = \frac{1}{5}$

#### Λύση

### Ερώτηση 1:

* Σχεδιάστε τις 1000 διαδρομές σε ένα γράφημα χρόνου-θέσης.
* Αυτές που έχουν τάση προς τα αριστερά με κόκκινο χρώμα και αυτές που έχουν τάση προς τα δεξιά με μπλέ 
* Σχολιάστε το αποτέλεσμα

#### Λύση

```python
fig, ax = plt.subplots()  # initiate plots
a= -70 # lower boundary
b=100 # upper boundary

#left
walk1_time_right =[]
N=500
for i in range(N):
    walk1 = rand_walk_Z(probs = (1/5,4/5))
    x =0 ;
    wa11=[x] 
    time=0 ;
    while (a<=x<=b) : 
        time+=1
        x = next(walk1)
        wa11.append(x)
    ax.plot(wa11,color = 'b',alpha =0.5)
    walk1_time_right.append(time)


#right
walk2_time_left =[]
for i in range(N):
    walk2 = rand_walk_Z(probs = (4/5,1/5))
    x =0 ;
    wa12=[x] 
    time=0 ;
    while (a<=x<=b) : 
        time+=1
        x = next(walk2)
        wa12.append(x)
    ax.plot(wa12,color = 'r',alpha =0.5)
    walk2_time_left.append(time)

    
##Better styling
plt.axhspan(a,b, color = '#cccccc', alpha = 0.8)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Απλός Συμμετρικός περίπατος στον $\mathbb{Z}$")
plt.grid()
plt.show()

    
```

Σχολιασμός του παραπάνω διαγράμματος:

Βλέπουμε  ότι  το γράφημα που περιέχει τους περιπάτους είναι συμμετρικο . Βέβαια, υπάρχει μια μικρή επικάλυψη των κόκκινων και των μπλε περιπάτων κοντά στο 0. Ωστόσο, οι περίπατοι τείνουν γρήγορα προς το άπειρο ($+\infty$ για τον μπλε και $-\infty$. για τον κόκκινο) και δεν υπάρχει επιπλέον επικάλυψη.

## Άσκηση 2:

Δημιουργήστε μία αλυσίδα που να κινείται στον $\mathbb{Z}_+$ με πιθανότητες μετάβασης:

* Για $k > 0$: $p(k, k+1) = \frac{1}{3}$ και $p(k, k-1) = \frac{2}{3}$
* Για $k = 0$: $p(0, 1) = \frac{1}{3}$ και $p(0, 0) = \frac{2}{3}$

Προσομοιώστε την αλυσίδα `1000` φορές.




### Ερώτηση 1:

Σχεδιάστε το ιστόγραμμα της θέσης $x_{100}$ όπου βρίσκεται η αλυσίδα μετά το εκατοστό βήμα της.



```python
x_100 = []
N=1000
M=100
for i in range(N):
    walk = rand_walk_Z2()
    x = 0
    for i in range(M):
        x = next(walk)
    x_100.append(x)
plt.hist(x_100, bins=30, edgecolor='black', density=True)
plt.xlabel('State')
plt.ylabel('Density')
plt.title('Histogram of the 100th position')
plt.grid()
plt.show()
```

### Ερώτηση 2:

Έστω:

$$ M_n = \max_{0\le k\le n} X_k$$

η δεξιότερη θέση που φτάνει η αλυσίδα στα $n$ πρώτα της βήματα.

1. Κάντε ιστόγραμμα της κατανομής της $M_{1000}$
2. Υπολογίστε τη μέση τιμή και τη διασπορά της.

```python
M_1000 = []
N=1000
M=1000
for i in range(N):
    walk = rand_walk_Z2()
    x = 0
    max1=0
    for i in range(M):
        x = next(walk)
        if (x >= max1):
            max1=x
    M_1000.append(max1)

plt.hist(M_1000, bins=30, edgecolor='black', density=True)
plt.xlabel('Position')
plt.ylabel('Density')
plt.title('Histogram of $M\_1000$ distribution')
plt.grid()
plt.show()
```

```python
print ("Η ζητούμενη μέση τιμή είναι :", stat.mean(M_1000) )
print ("Η ζητούμενη διασπορά είναι :", stat.variance(M_1000) )
```

## Άσκηση 3
 
Για τον τυχαίο περίπατο στο $\mathbb{Z}^2$

Προσομοιώστε 500 αλυσίδες και υπολογίστε το μέσο χρόνο που κάνει μια αλυσίδα για να βγει από τον δίσκο ακτίνας:

$$ R = 10, 20, ..., 100 $$

* Φτιάξτε ένα διάγραμμα $R$ - χρόνου εξόδου σε λογαριθμική κλίμακα.
* Σχολιάστε τη σχέση των 2 μεταβλητών

Στη συγκεριμένη άσκηση εκτός από μικρότερο αριθμό δειγμάτων είναι καλό να δοκιμάσετε τον κώδικα σας και για μικρά (~50) R στην αρχή καθώς η αλυσίδα μπορεί να χρειαστεί αρκετό χρόνο να ξεφύγει για μεγάλα R.

#### Λύση

```python
np.random.seed(2021)  # for reproducibility
state = [[ 0, 1 ], [ 0, -1 ], [ 1, 0 ], [ -1, 0 ] ]
list1=[]
means = []
M=500
rad=range( 10, 101, 10)
for R in rad:
    count=[]
    for i in range(M):
        time=0
        x = 0
        y=0
        while(x**2 + y ** 2 <= R ** 2 ):
            check = state[np.random.randint( 4 )]
            x += check[ 0 ]
            y += check[ 1 ]
            time += 1
        count.append(time)
    list1.append(R)
    mean_value= stat.mean(count)
    means.append(mean_value)

fig, ax = plt.subplots()  # initiate plots 
plt.loglog(list1,means,base=2)
plt.title( "Average Escape Time at given radius")
plt.xlabel( "Radius" )
plt.ylabel( "Escaping Time " )
plt.show()
```

```python
a,_ = np.polyfit(np.log2(list1),np.log2(means),1)
print ("Η κλίση της ευθείας είναi {0}".format(a))
```

## Σχολιασμός Αποτελεσμάτων 

Ισχύει  $ \frac{ log_2 ( T ) }{ \log_2 ( R ) } = 2$, αφου η σχέση χρόνου διαφυγής και ακτίνας είναι γραμμική σε λογαριθμική κλίμακα. Η παράμετρος 2 προσδιορίστηκε με την χρήση της γραμμικής παρεμβολής. Αυτό σημαίνει ότι η σχέση της ακτίνας με τον χρόνο φυγής είναι τετραγωνική ($ T= R^2 $).
