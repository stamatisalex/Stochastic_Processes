## Παραδοτέο 1

```python
import random
import numpy as np
import statistics as stat
from simple_markov_chain_lib import markov_chain
random.seed(2021)  # for reproducibility


p_list=np.arange( 0.02,1.0, 0.02 ) #list of all probabilities
total_list=[]
for p in p_list:
    initial_probs = {"0-0": 1.0}
    q=1-p
    markov_table = {
        "0-0": { "15-0": p,"0-15": q},
        "15-0": { "30-0": p,"15-15":q},
        "0-15": {"15-15": p,"0-30": q },
        "30-0": { "40-0": p,"30-15": q },
        "15-15": { "30-15": p,"15-30": q },
        "0-30": { "15-30": p,"15-40": q },
        "40-0": { "GameA": p,"40-15": q },
        "30-15": { "40-15": p,"Deuce": q },
        "15-30": { "Deuce": p,"15-40": q },
        "0-40": { "15-40": p,"GameB": q },
        "40-15": { "GameA": p,"AdvA": q },
        "Deuce": { "AdvA": p,"AdvB": q },
        "15-40": { "AdvB": p,"GameB": q},
        "GameA": { "GameA": 1.0 },
        "AdvA": { "GameA": p,"Deuce": q },
        "AdvB": { "Deuce": p,"GameB": q},
        "GameB": { "GameB": 1.0 },
    }

    mc = markov_chain( markov_table, initial_probs)

    #Experiment 
    N=20000
    list_track=[]
    count=0
    for i in range(10):
        for i in range(N):
            mc.start() 
            while mc.running_state not in ["GameA","GameB"]:
                mc.move()
            if mc.running_state=="GameA":
                count += 1
        mean=count/N
        list_track.append(mean)
        count=0
    result= np.mean(list_track)
    total_list.append(result)
    list_track.clear()
plt.figure(figsize=(12, 7))
plt.subplot()
plt.plot(p_list,total_list)
plt.xlabel("Πιανότητα p ")
plt.ylabel("Πιθανότητα να κερδίσει το παιχνίδι")
plt.title("Πιθανότητα νίκης του παίκτη που σερβίρει συναρτήσει της πιθανότητας p")
plt.grid(True)
plt.show()
```

### Άσκηση 

Σε αυτή την άσκηση θα μελετήσουμε την διακύμανση του χρόνου απορρόφησης μιας αλυσίδας Markov.
Στο παρακάτω κελί δημιουργούμε μια αλυσίδα Markov που ζει στο χώρο καταστάσεων $\mathbb{X} = \{0, 1, 2, 3, 4\}$ με πίνακα πιθανοτήτων μετάβασης

$$
P = 
\begin{pmatrix}
2/5   & 1/2 & 1/10 &  0  & 0 \\
1/3 &  0  &  0  & 2/3 & 0 \\
0   &  0  &  1  &  0  & 0 \\
1/2 &  1/8  &  0  & 1/4 & 1/8 \\
0   &  0  &  0  &  0  & 1 
\end{pmatrix}
$$

και αρχική κατανομή: 

$$ \pi_0 = (1, 0, 0, 0, 0 ) $$

Ο κώδικας της άσκησης είναι ο εξής:


```python
import random
import statistics as stat
import matplotlib.pyplot as plt 
import numpy as np

from simple_markov_chain_lib import markov_chain  # import markov chain simulator

%matplotlib inline

# Transition Table
markov_table = {
    0: {1: .5, 2: .5},  # from state 0 we move to state 1 with prob 0.5 and to state 2 with 0.5
    1: {0: 1/3, 3: 2/3},
    2: {2: 1.},
    3: {0: .5, 3: .25, 4: .25},
    4: {4: 1.}
}

# Initial Distribution
init_dist = {0: 1.}  # we start from state 0 with probability 1

M = 30 #we run the experiment 30 times for every sample_size

mc = markov_chain(markov_table, init_dist)


list_sam = []
list_var = []
N = np.arange(2**5,2**12,4)
for sample_size in N:
    total_list = []
    for runs in range( M ):
        running_total = 0
        for i in range(sample_size):
            mc.start()
            while mc.running_state != 2 and mc.running_state != 4:
                mc.move()
            running_total += mc.steps  # steps it took to be absorbed

        mc_estimate = running_total / sample_size
        total_list.append( mc_estimate )
        
    list_var.append(stat.variance( total_list ))
    list_sam.append(sample_size)
    
plt.figure(figsize=(12, 7))  # define figure size
```

```python
# Right Axes
plt.plot(N, list_var)
plt.xlabel('N')
plt.ylabel('Var(N)')
plt.title('Linear Scale')
plt.grid(True)  # add grid-lines
plt.show()

```

```python
(a,b) = np.polyfit(np.log2(N),np.log2(list_var),1)

Y = a * np.log2(N) + b

print('Η ζητούμενη κλήση της ευθείας είναι: {0}'.format(a))
plt.plot(np.log2(N), np.log2(list_var))
plt.plot(np.log2(N),Y)
plt.xlabel('logN')
plt.ylabel('Var(N)')
plt.title('Log-Log Scale')
plt.grid(True)
plt.show()
```

Άρα με βάση τα παραπάνω βλέπουμε ότι η ζητούμενη κλίση της ευθείας που εμφανίζεται είναι: -1.0146069353147707e-05

Παρατηρούμε ότι η γραφική παράσταση συμφωνεί με την θεωρία καθώς η διασπορά των εκτημίσεων μειώνεται με την αύξηση των δειγμάτων. Συγκεκριμένα έστω $Y_i = ( X_i - μ ) ^ 2$  ανεξάρτητες, ισόνομες τυχαίες μεταβλητές με μέση τιμή $\mathbb{E} [ Y_i ] = \mathbb{E} [ ( X_i - μ ) ^2 ] = \sigma _X ^2$. 

Αν $Var(x) = \frac{1}{N}\sum_{i=1}^{N} (X_i - μ)^2$, τότε $Var(x) =\frac{1}{N} \sum_{i=1}^{N} Y_i  = \frac{\sqrt{N}}{\sqrt{N}} ( \frac{1}{N} \sum_{i=1}^{N} Y_i - μ_Υ ) + μ_Υ $.

Από Κ.Ο.Θ έχουμε 

$Var(x) \to \frac{1}{\sqrt{N}} \mathcal{N}(μ_Υ, σ_Υ ^ 2) = \mathcal{N} ( \frac{\mu_Y}{\sqrt{N}}, \frac{\sigma_Y ^2}{N^{1/4}} )$ 

Μας ενδιαφέρει η ασυμπτωτική συμπεριφορά της εκτιμήτριας για $Ν \to \infty$, όπου οι τυχαίες μεταβλητές ακολουθούν κανονική κατανομή με μέση τιμή  $\frac{ \mu_Y }{\sqrt{N}}$. Όμως η μέση τιμή της διασποράς ουσιαστικά συμπεριφέρεται σαν $\frac{1}{\sqrt{N}}$ αφου το $\mu_Y$ είναι σταθερά.

Έχουμε δηλαδή $Var(X) = \Theta( \frac{1}{\sqrt{N}} )$ πράγμα που αποδείχθηκε και πειραματικά παραπάνω.
