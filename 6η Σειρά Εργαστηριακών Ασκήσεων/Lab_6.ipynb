Γράψτε μια ρουτίνα που θα τυπώνει για διαστάσεις $d=2,3,...$ τον εκτιμόμενο με Monte Carlo όγκο της μοναδιαίας μπάλας σε $d$ διαστάσεις, την πραγματική τιμή του και το σχετικό σφάλμα της εκτίμησης. Η ρουτίνα σας θα πρέπει να επιλέγει $N=10^6$ σημεία στον αντίστοιχο κύβο κάθε διάστασης και θα πρέπει να φτάνει μέχρι εκείνη την τιμή της διάστασης που κανένα από τα $N$ επιλεγμένα σημεία δεν έπεσε στη μπάλα. 


*Λύση*

Ο κώδικας της άσκησης  για d =2 ,...,20 είναι ο εξής: 

```python
from numpy import random
from math import gamma, pi

list1 = []
list2 = []

N = 10 ** 6
for d in range( 2,21):
    nhits = 0
    for i in range(N):
        x = random.uniform(-1,1,d)
        if sum(x ** 2) < 1: 
            nhits += 1
    vol= 2 ** d * nhits / N
    x = d/2
    vol_real= pi ** x / gamma(x + 1)
    list1.append(vol)
    list2.append(vol_real)

    print ("The Monte Carlo estimate of ω(", d,") is:\t%.5f " % vol )
    print("The actual value of ω(", d,") is:\t\t%.5f " % vol_real)
    print ("Relative error in ", d, " dimensions is:\t%.5f \n" %(abs(vol - vol_real)/vol_real))
```

Κάντε τη γραφική παράσταση $d- \log p(d)$ για τη θεωρητική τιμή της πιθανότητας $p(d)$. Εξηγήστε γιατί το σχετικό σφάλμα της εκτίμησης στο (α) είναι μεγάλο καθώς η διάσταση αυξάνει.

*Λύση*

```python
import random
import statistics as stat
import matplotlib.pyplot as plt 
import numpy as np
from math import log

%matplotlib inline

def Vol1(d):
    x = d/2
    return pi ** x / gamma(x + 1)

l1 = []
for i in range(2,21):
    l1.append(Vol1(i)/2**i)

logp = [np.log(x) for x in l1]
d = np.arange(2,21)
plt.figure(figsize=(12, 7))
plt.plot(d,d-logp);
plt.xlabel("Dimension");
plt.ylabel("d-Log(p(d))");

plt.figure(figsize=(12, 7))
plt.plot(d,logp);
plt.xlabel("Dimension");
plt.ylabel("Log(p(d))");
```

Παραπάνω σχεδιάστηκαν και η γραφική παράσταση "d-logp(d)" και η "logp(d)".
Παρατηρούμε ότι το σχετικό σφάλμα είναι μεγάλο για τις μεγάλες διαστάσεις. Αυτό οφείλεται στο γεγονός ότι σε μεγάλες διαστάσεις, το ποσοστό του μοναδιαίου κύβου που καταλαμβάνει η εγγεγραμμένη σε αυτόν μπάλα είναι πολύ μικρό. ## Παραδοτέο 2

α) Κατασκευάστε ένα scatter plot στο οποίο θα φαίνονται τα 1000 δείγματα που πήρατε απο τον δίσκο $D_2$ με τον παραπάνω αλγόριθμο MCMC. Στη συνέχεια κάντε το ίδιο για delta = 0.01 και για delta = 20. Εξηγήστε σε ένα κελί κειμένου γιατί το αποτέλεσμα 'χαλάει' όταν το delta είναι "πολύ μικρό" ή "πολύ μεγάλο"?



```python
import matplotlib.pyplot as plt
import random as r

for delta in[1.0 ,0.01,20.0]:
    N = 100 # number of steps to approach equilibrium
    samples = 1000 
    point_x = []
    point_y = []

    for _ in range(samples):
        x = [0,0]  ## start at the centre of th disc. This variable will keep the position of the chain
        R_sq = 0.0  ## this variable keeps the squared distance from 0. It saves some computations to keep it

        for _ in range(N):
            k = r.choice([0,1])  ## choose a jump direction at random
            z = r.uniform(-delta,delta) ## choose a jump size uniformly in (-delta,delta)
            x_prop_k = x[k] + z   ## propose a jump by z in the direction k
            R_sqprop = R_sq - x[k]**2+ x_prop_k**2 ## compute the squared distance from 0 after the proposed jump 
            if R_sqprop < 1.0: 
                R_sq = R_sqprop
                x[k]= x_prop_k   ## if the proposed jump leads to a point in the disc, then jump

        point_x.append(x[0])
        point_y.append(x[1])

    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('For delta ={}'.format(delta))
    plt.rcParams['figure.figsize'] = [6,6];
    plt.scatter(point_x,point_y);
    plt.show();
```

Παραπάνω βλέπουμε τα 3 διαγράμματα για delta=1 , delta =0.01 και delta =20 αντίστοιχα.Στην  περίπτωση που delta =0.01 τα δείγματα δεν "προλαβαίνουν" να απομακρινθούν σε 100 βηματα αρκεά από το κέντρο, αφού όπως βλέπουμε ελάχιστα ξεπερνούν την απόσταση του 0.1 από το (0,0). Αντιθέτως, στην δεύτερη περίπτωση οι κινήσεις πολλών δειγμάτων απορρίπτονται με αποτέλεσμα τα δείγματα αυτά να μην φεύγουν από  τους άξονες x ή y και να συγκεντρώνονται σε αυτούς οπως βλέπουμε. 


##### Ερώτημα β

Γράψτε μια ρουτίνα που θα τυπώνει για διαστάσεις $d=2,3,...,100$ τον εκτιμόμενο με MCMC όγκo της μοναδιαίας μπάλας σε $d$ διαστάσεις, την πραγματική τιμή του και το σχετικό σφάλμα της εκτίμησης. Μπορείτε να ξεκινήσετε από την προφανή σχέση $|D_1|=2$ και να δουλέψετε επαγωγικά για να υπολογίσετε τα ω(2), ω(3),... Στο jupyter notebook που θα παραδώσετε τρέξτε τη ρουτίνα σας ώστε να φαίνονται τα αποτελέσματα.


*Λύση*

Ξέρωντας ότι είναι $|D_1| = 2$, οπότε μέσω  της  MCMC δειγματοληψίας εκτιμούμε τον λόγο $\frac{|D_{d+1}|}{|C_{d+1}|}$= $\frac{|D_{d+1}|}{2|D_d|}$  που είναι ο λόγος των σημείων του κυλίνδρου σε d+1 διαστάσεις που πέφτουν μέσα στην d+1-διάστατη σφαίρα. Έτσι εκτιμούμε  το $|D_{d+1}|$ ενώ ξέρουμε το $|D_d|$. Ο κώδικας της άσκησης είναι ο εξής:

```python
import matplotlib.pyplot as plt
import random as r
import numpy as np


delta = 1
N = 100
lastd = 2
samples =10000     
for d in range(1,30): 
    samples_inside_D = 0 

    for _ in range(samples):
        x=[0]*d
        R_sq = 0 

        for _ in range(N): 
            k = r.choice(np.arange(d))
            z = r.uniform(-delta,delta)
            x_prop_k = x[k] +z
            R_sqprop = R_sq - x[k]**2 + x_prop_k**2
            if R_sqprop < 1.0: 
                R_sq = R_sqprop
                x[k]= x_prop_k

        x_dplus1 = r.uniform(-1,1)
        R_sq += x_dplus1**2

        if R_sq<1: 
            samples_inside_D += 1


    D=2*lastd*(samples_inside_D/samples)
    lastd=D
    error= abs(Vol1(d+1)-D)/Vol1(d+1)
    print("The Monte Carlo estimate of ω(", d+1 ,") is:\t%.5f " % D)
    print("The actual value of ω(", d+1 ,") is:\t\t%.5f " % Vol1(d+1))
    print("The relative error is:\t\t\t%.5f" % error)
    print("\n")
        
```
